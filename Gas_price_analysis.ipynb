## Import required libraries 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
```

## Load and Read file

```python
gas = pd.read_excel("Gas_prices.xlsx")
gas
```

## 1. In which country gas prices are highest?

```python
df1 = ['Australia','Mexico','USA','UK','Italy','Japan','Canada','France','Germany','South korea']
gas[df1].mean(axis=0).reset_index()
```

### Thus, averagely Italy saw the highest gas prices while USA has the cheapest gas prices averagely.

## 2.Which Year was the most expensive in terms of gas prices?

```python
gas[df1].mean(axis=1).reset_index()
```

### Thus, 2008 was averagely the most expensive year for every country in terms of gas prices.

```python
gas[df1].max().reset_index()
```

## 2. Comparison of gas prices between mexico and Italy over the years

```python
df2 = gas.plot(x="Year",y=["Mexico","Italy"], kind="bar" , rot=80)
```

```python
plt.figure(figsize=(8,8))
plt.title('ATTENDENCE OF BOYS IN MESS ON DIFFERENT DAYS')
plt.plot(gas.Year, gas.Mexico,'b.-',label='Mexico')
plt.plot(gas.Year, gas.Italy,'g.-',label='Italy')
plt.legend()
plt.show()
```
